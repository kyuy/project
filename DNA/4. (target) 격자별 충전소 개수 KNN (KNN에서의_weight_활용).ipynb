# KNN에서의 weight를 활용하여 격자별 충전소 개수 구하기

```python
import pandas as pd
import numpy as np

KNN_seoul_data_4 = pd.read_csv('KNN_seoul_data_4.csv')
KNN_seoul_data_4 = KNN_seoul_data_4.drop(columns='Unnamed: 0')
KNN_seoul_data_4
```

```python
KNN_seoul_data_4 = KNN_seoul_data_4.drop(columns='충전소명')
KNN_seoul_data_4
```

```python
a=KNN_seoul_data_4[['grid0','weight0']]
a
```

```python
b=KNN_seoul_data_4[['grid1','weight1']].rename(columns={'grid1':'grid0', 'weight1':'weight0'})
b
```

```python
c=KNN_seoul_data_4[['grid2','weight2']].rename(columns={'grid2':'grid0', 'weight2':'weight0'})
c
```

```python
d=KNN_seoul_data_4[['grid3','weight3']].rename(columns={'grid3':'grid0', 'weight3':'weight0'})
d
```

```python
result = pd.concat([a, b, c, d],ignore_index=True)
```

```python
result = result.groupby(['grid0']).agg({'weight0':'sum'})
result
```

```python
result = result.reset_index()
result
```

```python
result['weight0'].sum()
```

```python
result.to_csv('격자당 충전소개수_KNN.csv')
```
