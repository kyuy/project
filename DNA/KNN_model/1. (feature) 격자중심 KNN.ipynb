## 1. 전기차 충전소 "통합 데이터 파일 (한전, 환경부, 서울시)" 을 활용하여 충전소 위치 위경도 데이터 생성

```python
import pandas as pd
import numpy as np
```

```python
seoul_data=pd.read_csv('seoul_data(1km).csv',encoding='cp949')
seoul_data = seoul_data.drop(columns='Unnamed: 0')
seoul_data
```

```python
import pandas as pd
import numpy as np

# 필요한 행 선택
seoul_data_station_xy = seoul_data[['gid', 'y', 'x']]

#중복되는 데이터 없애기
station_xy = seoul_data_station_xy.drop_duplicates()

station_xy = station_xy.rename(columns={'y':'위도'})
station_xy = station_xy.rename(columns={'x':'경도'})

station_xy
```

```python
# 위치 정보 결측데이터 확인
station_xy[station_xy['경도'].isnull()==True]
```

```python
# 위치 정보 결측데이터 삭제
index_null = station_xy[station_xy['경도'].isnull()==True].index
station_xy = station_xy.drop(index = index_null)
station_xy
```

```python
station_xy.reset_index(drop=True)
```

## 2. 서울 교통량 데이터 다운로드 및 각 교통량 위치 포인트 중심의 위경도 데이터 생성

```python
traffic_data=pd.read_csv('traffic.csv',encoding='cp949')
traffic_data
```

```python
# 지점명칭별 경도 위도 생성 후 중복제거
traffic_location = traffic_data[['지점명칭','경도', '위도']].copy()
traffic_location.drop_duplicates(inplace=True, ignore_index=True)
traffic_location = traffic_location.rename(columns={'경도':'x','위도':'y'})
traffic_location
```

## KNN 수행, K=1 (교통량 데이터)

```python
# KNN의 K
K=1

KNN_traffic_data_1 = station_xy['gid'].to_frame().reset_index(drop=True)
for i in range(K) :
    KNN_traffic_data_1['지점명칭'+str(i)]=np.nan
    KNN_traffic_data_1['weight'+str(i)]=np.nan

KNN_traffic_data_1
```

```python
# KNN 수행하여 근접 1개 지점명칭과 가중치 출력 후 데이터 프레임에 넣기

for i in range(len(station_xy)):

    # 각 교통량 포인트와 충전소의 거리정보를 저장할 임시 데이터 프레임 dists를 만듬
    dists = traffic_location.copy()
    
    stationY=station_xy.iloc[i,1]    # y:위도
    stationX=station_xy.iloc[i,2]    # x:경도
      
    #새로운 데이터의 좌표와 알고있는 모든 데이터 집합의 좌표 사이 거리 계산
    dists['위도차'] = dists['y'] - stationY
    dists['경도차'] = dists['x'] - stationX
    
    dists['거리'] = dists['경도차']**2 + dists['위도차']**2
    
    #print(dists)
    
    dists.sort_values(by='거리', axis=0, ascending=True, inplace=True)
    
    #print(dists)
    
    select_gid = dists.iloc[0:K,:].copy()
    select_gid['거리역수'] = 1/select_gid['거리']
    select_gid['가중치'] = select_gid['거리역수']/sum(select_gid['거리역수'])
    
    #print(select_gid)
    
    select_gid = select_gid[['지점명칭', '가중치']]
    
    #print(select_gid)
    
    for j in range(K):
        KNN_traffic_data_1.iloc[i,2*j+1]=select_gid.iloc[j,0]
        KNN_traffic_data_1.iloc[i,2*j+2]=select_gid.iloc[j,1]
    
    #print(KNN_traffic_data_1)    
    
KNN_traffic_data_1
```

```python
# 데이터가 모두 잘 들어갔는지 확인
KNN_traffic_data_1.loc[KNN_traffic_data_1['weight0'].isnull()==True]
```

```python
KNN_traffic_data_1.to_csv('KNN_traffic_data_격자중심점기준_1.csv')
```
